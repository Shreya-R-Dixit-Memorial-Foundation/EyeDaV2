```python
import pandas as pd
data = pd.read_csv("acc.csv")
features = data[['XCOORD', 'YCOORD']]
print(features)
```

```python

from sklearn.cluster import KMeans
kmeans = KMeans(
    init="random",
    n_clusters=2000,
    n_init=10,
    max_iter=300,
    random_state=42
)

kmeans.fit(features)
labels = kmeans.labels_

data['cluster'] = labels
_clusters = data.groupby('cluster')['cluster'].count()
print(_clusters)
```

```python
centroids = kmeans.cluster_centers_
print(centroids)
print(len(centroids))
print(type(centroids))
print(type(data.iterrows()))
```

```python
import folium

map = folium.Map(location=[centroids[0][0], centroids[0][1]], zoom_start=12)
for centroid in centroids:
    folium.CircleMarker(
        location=[centroid[0], centroid[1]],
        radius=1, 
        weight=1, 
        fill=True, 
        fill_color="black",
        color="black"
    ).add_to(map)
map
```
